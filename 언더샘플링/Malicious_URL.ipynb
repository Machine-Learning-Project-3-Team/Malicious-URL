```python
import pandas as pd
```

```python
# CSV 파일 불러오기
df = pd.read_csv('cleaned_url_dataset.csv')
```

```python
# 'label' 컬럼에서 1 또는 2인 행만 필터링
malicious_df = df[df['label'].isin([1, 2])]
```

```python
# 결과 저장
malicious_df.to_csv('malicious_only.csv', index=False)
```
